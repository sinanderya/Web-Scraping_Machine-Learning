```python
# Importing Libraries

from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
url ="https://www.arabam.com/"
html = urlopen(url)
response = requests.get(url)
if response.status_code != 200 :
    print ('Hatalı/acılamayan sayfa ->', url)
else :
    print ('Response Status :', response.status_code)
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
print(soup.prettify())
```

```python
# find the title of the url 
def find_title():
    return soup.title.text

title = find_title()
print ('Title :', title)
```

Araç ilanlarını 86 il genelinde, link bazında topluyoruz.

```python
url = "https://www.arabam.com/ikinci-el/otomobil?city="
links = []
for j in range(86):
    for i in range(50):
        url1 = url + str(j+1) + "&take=50&page=" + str(i+1)
        html = urlopen(url1)
        soup = BeautifulSoup(html, "html.parser")  
        href = [div.a.get('href') for div in soup.find_all(class_='pr10 fade-out-content-wrapper')]
        links.append(href)
        # print('link-', links)
```

```python
# 2 boyutlu listeden tek boyutlu listeye geçtik.
links = [j for sub in links for j in sub]
```

Verilen her araç ilanı için url oluşturuyoruz.

```python
link = []
for i, value in enumerate(links):
    get_url = "https://www.arabam.com" + value
    link.append(get_url)
link
```

```python
df=pd.read_csv(r'C:\Users\Lenovo\Documents\GitHub\Web_Scraping & Machine_Learning\link_data.csv')
```

```python
df.rename(columns = {'0': 'a'}, inplace=True)
```

```python
cars_data = []

for i, value in enumerate(df.a):
    html = urlopen(value)
    soup = BeautifulSoup(html,'html.parser')
    
    row =[]
    
    # price
    price = soup.find('span', {'class' :'color-red4 font-semi-big bold fl w66'})
    if price == None:
        price = soup.find('span', {'class': 'color-red4 font-semi-big bold fl'})
        if price == None :
            row.append('None')
        else :
            row.append(price.text)
    else:
        row.append(price.text)
        
    # city
    city = soup.find(class_='one-line-overflow font-default-minus pt4 color-black2018 bold')
    if city == None :
        row.append('None')
    else:
        row.append(city.text)
        
    # info 
    info = soup.find('ul',{'class':'w100 cf mt16'})
    if info == None :
        row.append('None')
    else :
        row.append(info.text)
    cars_data.append(row)
```

İşlemler uzun sürdüğü için topladığımız bilgileri csv formatında excel alıyoruz.Burdan devam edeceğiz.

```python
df1 = pd.DataFrame(data=cars_data)
```
